```python
import requests
from bs4 import BeautifulSoup
response = requests.get("https://www.thefactsite.com/100-space-facts/")
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

```python
facts = soup.findAll('h2', {'class':'list'})
facts[:3]
print(len(facts)) # should be 100
```

```python
fact_list = []
for fact in facts:
    fact_list.append(fact.text)
    print (fact.text)
```

```python
facts_df = pd.DataFrame(data=fact_list, columns=["facts"])
facts_df.to_csv(r'Space_Facts.csv', index=False)
```
